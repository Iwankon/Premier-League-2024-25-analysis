```python
import pandas as pd
df = pd.read_csv('E0.csv')
df.head()
```

```python
df.shape
```

```python
matches = df[['Date','HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']]
matches.head()
```

```python
matches['FTR'].value_counts()
```

```python
home_points_map = {'H' : 3, 'D' : 1, 'A' : 0}
matches['home_points'] = matches['FTR'].map(home_points_map)
matches.head()
```

```python
away_points_map = {'H' : 0, 'D' : 1, 'A' : 3}
matches['away_points'] = matches['FTR'].map(away_points_map)
matches.head()
```

```python
home_pts = matches.groupby('HomeTeam')['home_points'].sum()
home_pts
```

```python
away_pts = matches.groupby('AwayTeam')['away_points'].sum()
away_pts
```

```python
total_pts = (home_pts + away_pts).sort_values(ascending=False)
total_pts
```
